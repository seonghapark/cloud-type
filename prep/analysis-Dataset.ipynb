## Mask images to remove building

```python
def mask_circle_solid(pil_img, background_color, blur_radius, offset=0):
    background = Image.new(pil_img.mode, pil_img.size, background_color)

    offset = blur_radius * 2 + offset
    mask = Image.new("L", pil_img.size, 0)
    draw = ImageDraw.Draw(mask)
    draw.ellipse((offset, offset, pil_img.size[0] - offset, pil_img.size[1] - offset), fill=255)
    mask = mask.filter(ImageFilter.GaussianBlur(blur_radius))

    return Image.composite(pil_img, background, mask)
```

```python
im_circle = mask_circle_solid(im, (0, 0, 0), 128)
```

```python
fig = plt.figure(figsize=(8,3))
ax = plt.subplot(1,2,1)
ax.imshow(im)
ax = plt.subplot(1,2,2)
ax.imshow(im_circle)
plt.show()
```

## Patch creation

```python
img = np.asarray(im)
```

```python
nx, ny = im.size
```

```python
patch_slide=128
patch_size=128
```

coords creation

```python
coords = []
for x in np.arange(0, nx, patch_slide):
    for y in np.arange(0, ny, patch_slide):
        if x + patch_size < nx and y + patch_size < ny:
            coords.append((x,y))
```

swath to patch

```python
patches = []
for i, j in coords:
    patches.append(img[i:i + patch_size, j:j + patch_size])
```

### Visual insepct

patch_slide=128, patch_size=128

```python
xsteps = np.arange(0, nx, patch_slide).shape[0]
ysteps = np.arange(0, ny, patch_slide).shape[0]
fig,axes = plt.subplots(xsteps, ysteps, figsize=(16,12))

idx = 1
for _patch, (ix, iy) in zip(patches, coords):
        #print(int(ix//128)+1, int(iy//128)+1, idx )
        #ax = plt.subplot(int(ix//128)+1, int(iy//128)+1, idx )
        ax = axes[int(ix//128), int(iy//128)]
        ax.imshow(_patch)
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        idx+=1
plt.show()
```

```python
xsteps = np.arange(0, nx, patch_slide).shape[0]
ysteps = np.arange(0, ny, patch_slide).shape[0]
patch_mode = np.zeros((xsteps, ysteps,3))
thres = 70
for i, j in coords:
    for k in range(3):
        # patch_mode[int(i//128), int(j//128), k] =
        v =  mode(
            img[i:i + patch_size, j:j + patch_size, k].ravel())[0][0] 
        #print(v)
        v = v if v > thres else 0.0
        patch_mode[int(i//128), int(j//128), k] =v
```

```python
# No threshold
plt.figure(figsize=(12,4))
for j in range(3):
    jj = j+1
    ax = plt.subplot(1,3,jj)
    c = ax.imshow(patch_mode[:,:,j])
    ax.set_title(f"BAND {jj}")
    plt.colorbar(c,ax=ax)
plt.show()
```

```python
# threshold 50
plt.figure(figsize=(12,4))
for j in range(3):
    jj = j+1
    ax = plt.subplot(1,3,jj)
    c = ax.imshow(patch_mode[:,:,j])
    ax.set_title(f"BAND {jj}")
    plt.colorbar(c,ax=ax)
plt.show()
```

## Visualize average statistic

```python
def calc_distribution(filelist):

    bands = defaultdict(list)
    for i in range(3):
        bands[i] = []
    
    for ifile in filelist:
        im = Image.open(ifile)
        img = np.asarray(im)
        for i in range(3):
            bands[i].extend(img[:,:,i].ravel() )
    return bands
```

```python
bands = calc_distribution(filelist[:200])
```

```python
len(bands[0])
```

```python
# Red-Green-blue
for i, v in bands.items():
    fig = plt.figure(figsize=(8,3))
    ax = plt.subplot(1,3,i+1)
    ax.hist(v, bins=range(250))
plt.show()
```
